```python
#Create a basket of all products, orders and quantity
basket17 = (ProductAttribute17.groupby(['OrderId','ProductCode'])['Quantity']).sum().unstack().reset_index().fillna(0).set_index('OrderId')
# .sum().unstack().reset_index().fillna(0).set_index('order_id')
basket17
```

```python
def convto0(x):
    if (x<=0):
        return 0
    if (x>=1):
        return 1
Basket_sets17=basket17.applymap(convto0)
Basket_sets17.head()
```

```python
from mlxtend.frequent_patterns import apriori ,association_rules
frequent_itemsets17 = apriori(Basket_sets17.astype('bool'), min_support = 0.0008, use_colnames = True)
frequent_itemsets17['length'] = frequent_itemsets17['itemsets'].apply(lambda x: len(x))
frequent_itemsets17
```

```python
lenght17=frequent_itemsets17['itemsets'].str.len()
Filter17=lenght17>=1  #1
frequent_itemsets17[Filter17]
```

```python
rules17 = association_rules(frequent_itemsets17, metric = 'support', min_threshold = 0.0006)
rules17
```
